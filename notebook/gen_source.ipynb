```python
import pandas as pd
```

```python
train_feature = pd.read_csv("task1_training_features.csv")
fix_feature = pd.read_csv("task1_negative_features_with_results.csv")
```

```python
train_feature.head(10)
```

```python
fix_feature.head(10)
```

```python
train_feature['predicted_rate'] = train_feature['rate']
```

```python
# 检查列名和列数差异
train_columns = set(train_feature.columns)
fix_columns = set(fix_feature.columns)

# 找出不一致的列
missing_in_train = fix_columns - train_columns
missing_in_fix = train_columns - fix_columns

print("不一致的列:")
print("train_feature 缺少的列:", missing_in_train)
print("fix_feature 缺少的列:", missing_in_fix)

# 获取并输出两个文件的所有列名
# print("train_feature 列名:", train_feature.columns.tolist())
```

```python
if list(train_feature.columns) == list(fix_feature.columns):
    # 合并 train_feature 和 fix_feature
    source = pd.concat([train_feature, fix_feature])
else:
    raise ValueError("train_feature 和 fix_feature 列数或列名不一致，无法合并")
```

```python
source = source.sort_values(by="original_index").reset_index(drop=True)
```

```python
# 对 'rate' 列进行处理，提取数值部分并转换为浮点数
source['rate'] = source['rate'].astype(str).str.extract(r'(\d+(\.\d+)?)')[0].astype(float)

# 对 'predicted_rate' 列进行相同处理
source['predicted_rate'] = source['predicted_rate'].astype(str).str.extract(r'(\d+(\.\d+)?)')[0].astype(float)

# 检查前几行，确认处理结果
print(source[['rate', 'predicted_rate']].head(10))
```

```python
# 统计 rate 列中非数值类型数据的数量，并输出前十个
non_numeric_rate_count = source[pd.to_numeric(source['rate'], errors='coerce').isna()].shape[0]
non_numeric_rate_samples = source[pd.to_numeric(source['rate'], errors='coerce').isna()].head(10)

# 统计 predicted_rate 列中非数值类型数据的数量，并输出前十个
non_numeric_predicted_rate_count = source[pd.to_numeric(source['predicted_rate'], errors='coerce').isna()].shape[0]
non_numeric_predicted_rate_samples = source[pd.to_numeric(source['predicted_rate'], errors='coerce').isna()].head(10)

# 打印结果
print("rate 列中非数值类型数据的数量:", non_numeric_rate_count)
print("rate 列中前 10 个非数值类型数据:")
print(non_numeric_rate_samples)

print("\npredicted_rate 列中非数值类型数据的数量:", non_numeric_predicted_rate_count)
print("predicted_rate 列中前 10 个非数值类型数据:")
print(non_numeric_predicted_rate_samples)
```

```python
# 将 rate 列中的所有空值或非数值类型数据替换为 0
source['rate'] = pd.to_numeric(source['rate'], errors='coerce').fillna(0)
```

```python
source['new_rate'] = (0.6 * source['rate'] + 0.4 * source['predicted_rate']).round(1)
source.to_csv("source.csv", index=False) 
print("第二步完成：已合并 train_feature 和 fix_feature，添加 new_rate 列，并生成 source.csv 文件")
```
